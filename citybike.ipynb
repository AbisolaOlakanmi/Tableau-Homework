```python
import pandas as pd
import numpy as np
import os
from pathlib import Path
```

```python
files18 = ['Resources/Data_2018/201801-citibike-tripdata.csv', 'Resources/Data_2018/201802-citibike-tripdata.csv', 'Resources/Data_2018/201803-citibike-tripdata.csv']
# Create an empty list to store each dataframe
df18_list = []

# Loop through each file in the list
for file in files18:
    # Read the file into a dataframe
    df18 = pd.read_csv(file)

    # Append the dataframe to the list
    df18_list.append(df18)

# Concatenate all dataframes in the list into one dataframe
df18 = pd.concat(df18_list)

# Show dataframe
df18
```

```python
files19 = ['Resources/Data_2019/201901-citibike-tripdata.csv', 'Resources/Data_2019/201902-citibike-tripdata.csv','Resources/Data_2019/201903-citibike-tripdata.csv']
# Create an empty list to store each dataframe
df19_list = []

# Loop through each file in the list
for file in files19:
    # Read the file into a dataframe
    df19 = pd.read_csv(file)

    # Append the dataframe to the list
    df19_list.append(df19)

# Concatenate all dataframes in the list into one dataframe
df19 = pd.concat(df19_list)

# Show dataframe
df19
```

```python
# Concatenate all dataframes for all the years into one
combined_df = pd.concat([df18, df19], axis=0)

# Display the results
combined_df
```

```python
# Dictionary mapping values in the 'gender' column to new values
mapping = {0: 'Unknown', 1: 'Male', 2: 'Female'}

# Use the 'map' method to change the values in the 'gender' column
combined_df['gender'] = combined_df['gender'].map(mapping)

# Show DataFrame
combined_df
```

```python
# Save the dataframe to a CSV file
combined_df.to_csv('Resources/citybike.csv', index=False)
```
